```python
import matplotlib.pyplot as plt

#membuatboxplot untuk kolom height dan weight
df.boxplot(column=['weight_kg', 'height_cm'])
plt.title('Boxplot of weight and height with outliner')
plt.show()
```

```python
# quartile 1
q1 = df['height_cm'].quantile(0.25)
print("Q1 : ",q1)

# quartile 3
q3 = df['height_cm'].quantile(0.75)
print("Q3 : ",q3)

iqr = q3 - q1
print("IQR : ",iqr)

#hitung batas bawah
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# temukan outlier dalam dataframe
outliner = df[(df['height_cm'] < lower_bound) | (df['height_cm'] > upper_bound)]

print("Outliner height_cm: ")
print(outliner)
```

```python
data_tinggi = df["height_cm"]
# Buat histogram
n, bins, patches = plt.hist(data_tinggi, bins=5, color='pink', edgecolor='black')

# Tambahkan Label
plt.title('Histogram Nilai')
plt.xlabel('Height')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, ['{:.0f}-{:.0f} ' .format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Tampilkan histogram
plt.show()
```

```python
# Mengidentifikasi baris outlier berdasarkan ciri-ciri yang unik

outliers_condition = ((df['height_cm'] > upper_bound) | (df['height_cm'] < lower_bound)) | (df['weight_kg'] < 50) | (df['weight_kg'] > 120)

# Drop baris yang memenuhi kondisi di atas (outliers)
df_cleaned = df[~outliers_condition]  # Tanda ~ untuk negasi (tidak termasuk kondisi outlier)

# Menampilkan data setelah penghapusan outliers
print(df_cleaned.tail())
```
